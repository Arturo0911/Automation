```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
figure = plt.figure()
axes = figure.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x,y, color = "red")
ax2 = fig.add_axes([0.2,.5,.2,.2])
ax2.plot(x,y)
```

```python
figure = plt.figure()
axes1 = figure.add_axes([0,0,1,1])
axes1.plot(x,z)
axes1.set_xlabel("x")
axes1.set_ylabel("z")

# chart insied
axes2 = figure.add_axes([0.2,0.5,.4,.4])
axes2.plot(x,y)
axes2.set_xlabel("x")
axes2.set_ylabel("y")
axes2.set_xlim([20,22])
axes2.set_ylim([30,50])


```

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y, color= "blue",lw=3,ls= "--")
axes[1].plot(x,z,color="red", lw=3, ls= "--")
```
