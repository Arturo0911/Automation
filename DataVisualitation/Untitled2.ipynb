```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
tips.head() # printing the first fours places
```

```python
sns.displot(tips['total_bill'],kde= False, bins = 100)
```

```python
sns.jointplot(x= "total_bill", y= "tip", data= tips)
# show a trend for the matchig data
```

```python
sns.pairplot(tips, hue= "sex", palette= "coolwarm") 
# hue parameters is not numeric, must to be categorycal
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.barplot(x="sex",y="total_bill",data=tips, estimator= np.std)
# set the x axis to categorycal and the other one numeric
```

```python
sns.countplot(x="sex", data= tips)
```

```python
sns.boxplot(x="day",y= "total_bill", data= tips, hue = "smoker")
```

```python
sns.violinplot(x = "day", y= "total_bill", data= tips, hue = "sex", split =True)
```

```python
sns.stripplot(x = "day", y= "total_bill", data= tips, jitter = True,split= True)
```

```python
sns.violinplot(x = "day", y = "total_bill", data = tips)
sns.swarmplot(x= "day", y = "total_bill", data= tips, color = "black")
```

```python
# using another function to general purpose
# using the comparation between information
# first one is categorical and the other one is numeric
sns.factorplot(x = "day", y= "total_bill", data = tips, kind= "violin")
```
